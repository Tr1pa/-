## 1.5 Группировка данных и визуальный анализ


```python
data.groupby(by='Age')['Crossing'].describe()
```

```python
Crossing = data[data['Age']>=20].boxplot(column='Crossing', by='Age')
```

```python
mean_20 = data[data['Age']>=20].groupby(by='Crossing')['Age'].mean()

mean_20

# инициализируем фигуру и указываем ее размер
plt.figure(figsize=(10, 6))

# название графика
plt.title('Изменение', fontsize=16)

# строим график
plt.plot(mean_20, color='green', linestyle='-.', label='график')

# настраиваем подписи осей
plt.ylabel('Age', fontsize=16)

# включаем легенду и сетку
plt.legend(fontsize=14)
plt.grid()
```

## 1.6 Заполним пропуски в данных

```python
data.info()
```

```python
data['Age'].value_counts()
```

```python
data['Nationality'].value_counts()
```

```python
data['Wage'].median()
```

```python
drive_filler = data['Age'].value_counts().index[0]
size_filler = data['Nationality'].value_counts().index[0]
weather_filler = data['Wage'].median()
```

Запрлним пропуски с помощью метода .fillna()

```python
data['Age'] = data['Age'].fillna(drive_filler)
data['Nationality'] = data['Nationality'].fillna(size_filler)
data['Wage'] = data['Wage'].fillna(weather_filler)
```
